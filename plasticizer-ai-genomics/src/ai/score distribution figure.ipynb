```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
os.chdir(r"C:\Users\Anuraj\Desktop\khushboo_backup\ai_ml_ksskslab")
abc= pd.read_csv("KSSKSLAB04_plasticizer_ranked.csv")
scores=abc["plasticizer_score"].astype(float)
BINS = 10
plt.figure(figsize=(7,5))
counts, bin_edges, patches= plt.hist(scores, bins=BINS, edgecolor="black", linewidth=0.7)
colors = plt.cm.tab10(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

labels = [f"{bin_edges[i]:.2f}–{bin_edges[i+1]:.2f}" for i in range(len(bin_edges) - 1)]
plt.legend(patches, labels, title="Score range", fontsize=8)

plt.xlabel("Predicted plasticizer hydrolase probability score")
plt.ylabel("Number of candidate proteins")
plt.title("Distribution of AI-predicted plasticizer hydrolase scores (KSSKSLAB04)")
plt.tight_layout()
plt.savefig("Score distribution figure", dpi=300, bbox_inches="tight")
plt.show()
```
